```python
#nltk => if user types [walk,walks,walkiing]=>then recommend in the form of the keyword walk 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    a=[]
    for i in text.split():
        a.append(ps.stem(i))
    return " ".join(a) 
```

```python
new_df['keyword']=new_df['keyword'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector_keyword)
similarity[5]
```

```python
x = 'the shawshank redemption'
new_df['MovieTitle'] == x
movie_index   = new_df[new_df['MovieTitle'] == x].index[0]
distances =  similarity[movie_index]
movies_list = sorted(list(enumerate(distances)), reverse =True, key=lambda x:x[1])[1:6]
x_movies = []
for i in  movies_list:
    x_movies.append(new_df.iloc[i[0]].MovieTitle)
#     print(new_df.iloc[i[0]].title)

x_movies
    
    
```
